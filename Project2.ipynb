```python
#Dictionary-Based Database Management System
database_dic = {}
num_record = int(input("Enter number of records: "))
def addNewData():
  for i in range(0, num_record):
    i = i + 1
    key = str(input("Enter a key #" + str(i) + ": "))
    value = float(input("Enter a value #" + str(i) + ": "))
    database_dic[key] = value
addNewData()
def viewAllData():
  print("The data records sorted by key to value is...")
  for key, value in database_dic.items():
    print("{", "'", key, "'", ":", value, "}")
  print("Total number of data records is", num_record)
viewAllData()
def search4DataRecords():
  key = str(input("Enter a key to search for its data record including key and value: "))
  if key in database_dic:
    print("'", key, "' key was found at " ,"{", "'", key, "'", ":", database_dic[key], "}")
  else:
    print("'", key, "' key was not found in the database")
search4DataRecords()
def updateORdeleteData():
  key = str(input("Enter a key to update or delete the value: "))
  if key in database_dic:
    print("'", key, "' key was found in database")
    value = float(input("Enter a value for '" + key + "' to delete the original value: "))
    database_dic[key] = value
    print("'", key, "' key has been updated to", "{", "'", key, "'", ":", database_dic[key], "}")
  else:
    while key not in database_dic:
      print("'", key, "' key doesn't exist, try again")
      key = str(input("Enter a key to update or delete the value: "))
    print("'", key, "' key was found in database")
    value = float(input("Enter a value for '" + key + "' to delete the original value: "))
    database_dic[key] = value
    print("'", key, "' key has been updated to", "{", "'", key, "'", ":", database_dic[key], "}")
    word = str(input("Do you want to update or delete any more dictionary? "))
    if word == 'yes':
      for i in range(0, num_record):
        i = i + 1
        key = str(input("Enter a key to update or delete the value: "))
        if key in database_dic:
          print("'", key, "' key was found in database")
          value = float(input("Enter a value for '" + key + "' to delete the original value: "))
          database_dic[key] = value
          print("'", key, "' key has been updated to", "{", "'", key, "'", ":", database_dic[key], "}")
        else:
          while key not in database_dic:
            print("'", key, "' key doesn't exist, try again")
            key = str(input("Enter a key to update or delete the value: "))
          print("'", key, "' key was found in database")
          value = float(input("Enter a value for '" + key + "' to delete the original value: "))
          database_dic[key] = value
          print("'", key, "' key has been updated to", "{", "'", key, "'", ":", database_dic[key], "}")
    elif word == 'no':
        print("You have decided to stop updating and deleting data in the dictionary")
    else:
      print("Enter either yes or no")
      word = str(input("Do you want to update or delete any more dictionary? "))
      while word != 'no' and 'yes':
        print("Enter either yes or no")
        word = str(input("Do you want to update or delete any more dictionary? "))
        if word == 'yes':
          key = str(input("Enter a key to update or delete the value: "))
          if key in database_dic:
            print("'", key, "' key was found in database")
            value = float(input("Enter a value for '" + key + "' to delete the original value: "))
            database_dic[key] = value
            print("'", key, "' key has been updated to", "{", "'", key, "'", ":", database_dic[key], "}")
          else:
            while key not in database_dic:
              print("'", key, "' key doesn't exist, try again")
              key = str(input("Enter a key to update or delete the value: "))
            print("'", key, "' key was found in database")
            value = float(input("Enter a value for '" + key + "' to delete the original value:"))
            database_dic[key] = value
            print("'", key, "' key has been updated to", "{", "'", key, "'", ":", database_dic[key], "}")
        elif word == 'no':
          print("You have decided to stop updating and deleting data in the dictionary")
updateORdeleteData()
```
